```python
def treat_trip_times(time):
    split = time.split(":", 2)[:3]
    converted_to_secs = (int(split[0]) * 60 * 60) + (int(split[1]) * 60) + (int(split[2]))
    
    if converted_to_secs <= 21600:
        return converted_to_secs + 86400  
    else:
        return converted_to_secs

#S = [s for s in D_frequencies['start_time']]
#L = [s for s in D_frequencies['end_time']]

#import itertools as it
#LT = [treat_trip_times(y) - treat_trip_times(x) for x,y in it.product(S, L)]     
#D_frequencies['trip_total_time'] = D_frequencies['stop_id'].map(stops_mapper)
#treat_trip_times(D_frequencies['end_time'][1]) - treat_trip_times(D_frequencies['start_time'][1])

D_frequencies['start_time_sec'] = D_frequencies['start_time'].map(treat_trip_times)
D_frequencies['start_time_sec'] = D_frequencies['start_time_sec'].astype(int)

D_frequencies['end_time_sec'] = D_frequencies['end_time'].map(treat_trip_times)
D_frequencies['end_time_sec'] = D_frequencies['end_time_sec'].astype(int)

D_frequencies['sec_range'] = D_frequencies['end_time_sec'] - D_frequencies['start_time_sec']


D_frequencies
```

```python
D_frequencies['headway_secs'] = D_frequencies['headway_secs'].astype(int) 
D_frequencies
D_frequencies['trip_id_total_trains'] = (round(D_frequencies['sec_range'] / D_frequencies['headway_secs']))

D_frequencies['trip_id_total_trains'] = D_frequencies['trip_id_total_trains'].astype(int) 

#D_frequencies['trip_id_total_trains'] = D_frequencies['headway_secs'] / D_frequencies['trip_id_total_time']
#D_frequencies['trip_id_total_trains'] = D_frequencies['trip_id_total_trains'].astype(int)

#triptrains_mapper = pd.Series(D_frequencies.trip_id_total_trains.values, index = D_frequencies.trip_id).to_dict()
#print(triptrains_mapper)

D_frequencies
```
